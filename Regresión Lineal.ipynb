```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
data = pd.read_csv("/Users/estebanquinteropedroza/Desktop/Data/data.csv",sep=",")
```

```python
data
```

```python
data.columns
data.info
data.describe
```

```python
data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metro = data["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metro, data["precio"])
```

```python
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)",modelo.coef_)
```

```python
Entrada = [[15],[25]]
predicciones = modelo.predict(Entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label= 'Datos Originales')
plt.scatter(Entrada, predicciones, color ='red')
plt.plot(Entrada, predicciones, color ='black',label= 'Línea de regresión')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
